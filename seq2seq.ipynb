```python
import torch
import torch.nn as nn
import torch.optim as optim
from seq2seq import Encoder, Decoder, Seq2Seq, train, evaluate, plot_losses, plot_attention
from seq2seq import *
```

```python
# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# Define hyperparameters
input_size = 1000  # Vocabulary size
output_size = 1000  # Vocabulary size
hidden_size = 256
learning_rate = 0.001
num_epochs = 10
batch_size = 32
```

```python
# Initialize models
encoder = Encoder(input_size, hidden_size).to(device)
decoder = Decoder(hidden_size, output_size).to(device)
model = Seq2Seq(encoder, decoder, device).to(device)
```

```python
# Define optimizer and loss function
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.CrossEntropyLoss()
```

```python
# Create dummy data
source_data = torch.randint(1, input_size, (100, 10)).to(device)
target_data = torch.randint(1, output_size, (100, 10)).to(device)

# Split data into train and validation sets
train_data = (source_data[:80], target_data[:80])
val_data = (source_data[80:], target_data[80:])
```

```python
# Training loop
train_losses = []
val_losses = []

for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    for i in range(0, len(train_data[0]), batch_size):
        batch_source = train_data[0][i:i+batch_size]  # Shape: (batch_size, seq_len)
        batch_target = train_data[1][i:i+batch_size]  # Shape: (batch_size, seq_len)
        loss = train(model, batch_source, batch_target, optimizer, criterion)
        epoch_loss += loss
    
    train_losses.append(epoch_loss / (len(train_data[0]) // batch_size))
    
    model.eval()
    val_loss = evaluate(model, val_data[0], val_data[1], criterion)
    val_losses.append(val_loss)
    
    print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_losses[-1]:.4f}, Val Loss: {val_losses[-1]:.4f}")

# Plot losses
plot_losses(train_losses, val_losses)
```

```python
# Create a simple vocabulary
vocab_size = 1000
vocab = create_vocab(vocab_size)

# Test the model
model.eval()
test_source = torch.randint(1, vocab_size, (3, 10)).to(device)  # Shape: (batch_size, seq_len)
test_target = torch.randint(1, vocab_size, (3, 10)).to(device)  # Shape: (batch_size, seq_len)

with torch.no_grad():
    output = model(test_source, test_target, 0)
    predicted = output.argmax(2)

# Convert numbers to words
source_words = numbers_to_words(test_source.cpu(), vocab)
target_words = numbers_to_words(test_target.cpu(), vocab)
predicted_words = numbers_to_words(predicted.cpu(), vocab)

# Print results
print("Source sentences:")
for sentence in source_words:
    print(sentence)

print("\nTarget sentences:")
for sentence in target_words:
    print(sentence)

print("\nPredicted sentences:")
for sentence in predicted_words:
    print(sentence)
```
